```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
import pickle
# new created evidence and inutition sets 
with open("../preprocessing_EMI/evidence_set_final.pkl", "rb") as f:
    evidence_set = pickle.load(f)

with open("../preprocessing_EMI/intuition_set_final.pkl", "rb") as g:
    intuition_set = pickle.load(g)
```

```python
intuition_set.remove('fake')
```

```python
sns.set()
plt.rcParams['figure.figsize'] = [20, 15]

x, y = np.ogrid[:300, :300]
mask = (x - 150) ** 2 + (y - 150) ** 2 > 130 ** 2
mask = 255 * mask.astype(int)

wordcloud = WordCloud(background_color="white", mask=mask, contour_width=0.1, 
                      contour_color="white",  max_font_size=150, random_state=42,
                      colormap="Purples")

wordcloud.generate(text=' '.join(intuition_set))   
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
